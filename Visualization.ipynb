```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
plt.close("all")
ts = pd.Series(np.random.randn(1000),index= pd.date_range("1/1/2000", periods = 1000))
ts = ts.cumsum()
ts.plot()
```

```python
df = pd.DataFrame(np.random.randn(1000,10),index = ts.index,columns = list("ABCDEFGHIJ"))
df = df.cumsum()
plt.figure()
df.plot()
df
```
